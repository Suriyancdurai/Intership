### 3) Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL.


```python
page=requests.get("https://www.dineout.co.in/delhi-restaurants")

page
```

```python
soup = BeautifulSoup(page.content)

soup
```

```python
Restaurant_name=[]

for i in soup.find_all('a',class_='restnt-name ellipsis'):
    Restaurant_name.append(i.text)
    
Restaurant_name  
```

```python
cuisine=[]

for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text)
    
cuisine
```

```python
location=[]

for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
    
location
```

```python
rating=[]

for i in soup.find_all('div',class_=["restnt-rating rating-3","restnt-rating rating-4","restnt-rating rating-5"]):
    rating.append(i.text)
    
rating
```

```python
image_url=[]

for i in soup.find_all("img",class_="no-img"):
    image_url.append(i['data-src'])
    
image_url 
```

```python
df = pd.DataFrame({"Restaurant_name ":Restaurant_name  ,"Cuisine":cuisine,"Location":location,"Rating":rating,"Image_Url":image_url})
```

```python
df
```
