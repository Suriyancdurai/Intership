### 5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps. 

```python
#connecting to thr webdriver
driver=webdriver.Chrome()
```

```python
#getting webpage
driver.get('https://www.google.co.in/maps')
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
```

```python
def get_coordinates(city):
    search_bar=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/form/input')
    search_bar.send_keys(city)
    search_bar.send_keys(Keys.RETURN)
    time.sleep(3)
    
    current_url=driver.current_url
    lat_lon=current_url.split('@')[1].split(',')[0:2]    #extract the coordinates from url
    latitude,longitude=float(lat_lon[0]),float(lat_lon[1]) 
    return latitude,longitude
    
```

```python
#using function entering 'bangalore' in the search bar
city_to_search='Bangalore'
latitude,longitude=get_coordinates(city_to_search)
print(f"Coordinates for {city_to_search}: latitude{latitude},longitude{longitude}")
```

```python
driver.close()
```

```python
ishow_speed_comments=[] #assigning empty lists for ishow_speed_comments
ishow_speed_comments_upvote=[] #assigning empty lists for ishow_speed_comments_upvote
ishow_speed_comments_time=[] #assigning empty lists for ishow_speed_comments_upvote
```

```python
#For ishow_speed_comments(using XPATH)
try:
    comments_tags=driver.find_elements(By.XPATH,'//yt-attributed-string[@id="content-text"]') 
    for i in comments_tags:
        comments=i.text.replace('\n','')
        ishow_speed_comments.append(comments)
except NoSuchElementException:
    ishow_speed_comments.append('Not present')
    
#For ishow_speed_comments_upvote(using XPATH)
try:
    comments_upvote=driver.find_elements(By.XPATH,'//span[@class="style-scope ytd-comment-engagement-bar"]')
    for i in comments_upvote:
        cu=i.text
        ishow_speed_comments_upvote.append(cu)
except NoSuchElementException:
    ishow_speed_comments_upvote.append('Not present')
        
    
#For ishow_speed_comments_time(using XPATH)
try:
    comments_time=driver.find_elements(By.XPATH,'//span[@class="style-scope ytd-comment-view-model"]/a')
    for i in comments_time:
        ci=i.text
        ishow_speed_comments_time.append(ci)
except NoSuchElementException:
    ishow_speed_comments_time.append('Not present')
       
```

```python
#Initializing a chromebrowser using webdriver

driver=webdriver.Chrome()
```

```python
#Entering location name(londan) using XPATH  

location_name=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[2]/div/div[1]/div/div[1]/div/div/div[2]/input')
location_name.send_keys("London")
```

```python
#clicking londan,England using XPATH

location_name_england=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[2]/div/div[1]/div/div[2]/div/ul/li[2]/button')
location_name_england.click()
```

```python
next_button=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[3]/div/div/button[2]/div')
next_button.click()
```

```python
lets_go_button=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[3]/div/div/button/div')
lets_go_button.click()
```

```python
#making empty list and fetching required data

hostel_name=[] 
distance_from_city_centre=[]
ratings=[]
total_reviews=[]
overall_review=[]
privates_from_price=[]
dorms_from_price=[]
product_description=[]
facilites=[]
```

```python
#For hostel_name (using XPATH)

hostel_tags=driver.find_elements(By.XPATH,'//div[@class="property-name"]/span')
for i in hostel_tags:
    hn=i.text
    hostel_name.append(hn)
    
#For distance_from_city_centre (using XPATH)    
distance_tags=driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
for i in distance_tags:
    dfcc=i.text
    distance_from_city_centre.append(dfcc)
    

#For ratings (using XPATH)    
ratings_tags=driver.find_elements(By.XPATH,'//span[@class="number"]')
for i in ratings_tags:
    rs=i.text
    ratings.append(rs)
    
ratings=ratings[2:33] #in website there is some featured hostel so limiting it with index



#For total_reviews (using XPATH)    
total_reviews_tags=driver.find_elements(By.XPATH,'//div[@class="review"]')
for i in total_reviews_tags:
    tr=i.text
    total_reviews.append(tr)
    
#For overall_reviews (using XPATH)    
overall_reviews_tags=driver.find_elements(By.XPATH,'//span[@class="keyword"]')
for i in overall_reviews_tags:
    ovr=i.text
    overall_review.append(ovr)
    
overall_review=overall_review[2:33]
```

```python
#getting url

london_hotel_url=[]


url=driver.find_elements(By.XPATH,'//div[@class="property-card"]/a[1]')
for i in url:
    london_hotel_url.append(i.get_attribute('href'))
    
london_hotel_url=london_hotel_url[2:33]
    
```

```python
london_hotel_url
```

```python
for i in london_hotel_url:
    driver.get(i)
    time.sleep(20)
    try:                                #fetching product_description
        prd=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[10]/div/div[2]/div[6]/div/div[2]')
        product_description.append(prd.text)
    except NoSuchElementException:
         product_description.append('-')
    try:                                #fetching privates_from_price
        pfp=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[12]/div/div/section/div/div/div[2]/div[1]/div[2]/div/div[2]/div[1]/div/div[1]/div[2]/div[1]')
        privates_from_price.append(pfp.text)
    except NoSuchElementException:
         privates_from_price.append('-')
    try:                                #fetching dorms_from_price
        dfp=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[12]/div/div/section/div/div/div[2]/div[2]/div[2]/div[1]/div[2]/div[1]/div/div[1]/div[2]/div[1]') 
        dorms_from_price.append(dfp.text)
    except NoSuchElementException:
         dorms_from_price.append('-')
    try:                                #fetching dorms_from_price
        fsg=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[14]/div/ul/li[4]/ul/li') 
        facilites.append(fsg.text)
    except NoSuchElementException:
         facilites.append('-')
```

```python
print(len(hostel_name),
len(distance_from_city_centre),
len(ratings),
len(total_reviews),
len(overall_review),
len(privates_from_price),
len(dorms_from_price),
len(product_description),
len(facilites))
```
