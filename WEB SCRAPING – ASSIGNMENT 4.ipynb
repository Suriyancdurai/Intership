### 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 

**A)Rank**

**B) Name** 

**C) Artist**

**D) Upload date** 

**E) Views** 


```python
#importing necessory libraries

import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
from selenium.common.exceptions import NoSuchElementException
import requests
```

```python
#Initializing a chromebrowser using webdriver

driver=webdriver.Chrome()
```

```python
#getting the webpage

driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
#Assigning empty lists to store most viewed youtube videos

Rank=[]
Name=[]
Artist=[]
Upload_date=[]
Views=[]
```

```python
#For Name(using XPATH)
Name_tags=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
for j in Name_tags:
    ne=j.text.split("[")[0]
    Name.append(ne)
    
#For Artist(using XPATH)
Artist_tags=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
for k in Artist_tags:
    ar=k.text
    Artist.append(ar)


#For Upload_date(using XPATH)
Upload_date_tags=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
for l in Upload_date_tags:
    ud=l.text
    Upload_date.append(ud)
    
#For Upload_date(using XPATH)
Views_tags=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
for m in Views_tags:
    vs=m.text
    Views.append(vs)

```

```python
print(len(Name),len(Artist),len(Upload_date),len(Views))
```

```python
df_for_most_viewed_videos=pd.DataFrame({'Rank':range(1,31),'Video Name':Name,'Artist':Artist,'Upload_date':Upload_date,'Views in billion':Views})
```

```python
#DataFrame for most viewed videos on YouTube from Wikipedia

df_for_most_viewed_videos
```

```python
driver.close()
```

### 4. Scrape the details of trending repositories on Github.com. Url = https://github.com/ You have to find the following details: 

**A)Repository title** 

**B)Repository description**

**C)Contributors count**

**D)Language used**


```python
#Initializing a chromebrowser using webdriver

driver=webdriver.Chrome()
```

```python
#getting the webpage

driver.get("https://github.com/")
```

```python
#Clicking explore menu using XPATH

explore_menu = driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div/div/div/aside/div[3]/div/a")
explore_menu.click()
```

```python
#Clicking trending option using XPATH

trending = driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/main/div[1]/nav/div/a[3]")
trending.click()
```

```python
#fetching url

trending_urls=[]
url=driver.find_elements(By.XPATH,'//a[@class="Link"]')
for i in url:
    trending_urls.append(i.get_attribute('href'))
```

```python
len(trending_urls)
```

```python
#Assigning empty lists to store  trending repository details

Repository_title=[]
Repository_description=[]
Contributors_count=[]
Language_used=[]
```

```python
for i in trending_urls:
    driver.get(i)
    time.sleep(5)
    try:                                #fetching Repository_title
        ryt=driver.find_element(By.XPATH,'//strong[@class="mr-2 flex-self-stretch d-none d-md-block"]') 
        Repository_title.append(ryt.text)
    except NoSuchElementException:
         Repository_title.append('-')
    try:                                #fetching Repository_description
        ryd=driver.find_element(By.XPATH,'//p[@class="f4 my-3"]')
        Repository_description.append(ryd.text)
    except NoSuchElementException:
        Repository_description.append('-')
    try:                                #fetching Contributors_count
        cct=driver.find_element(By.XPATH,'/html/body/div[1]/div[6]/div/main/turbo-frame/div/div/div/div[2]/div[2]/div/div[5]/div/h2/a/span')
        Contributors_count.append(cct.text)
    except NoSuchElementException:
        Contributors_count.append('-')
    try:                             #fetching Language_used
        lu=driver.find_element(By.XPATH,'//span[@class="color-fg-default text-bold mr-1"]')
        Language_used.append(lu.text)
    except NoSuchElementException:
        Language_used.append('-') 
```

```python
print(len(Repository_title),len(Repository_description),len(Contributors_count),len(Language_used))
```

```python
df_for_trending_repository=pd.DataFrame({'Repository_title':Repository_title,'Repository_description':Repository_description,
                                         'Contributors_count':Contributors_count,'Language_used':Language_used})
```

```python
#DataFrame for trending repositories on Github.com

df_for_trending_repository
```
